```python
#Import Libraries

import time

from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
#Import URL
website = 'https://www.adamchoi.co.uk/overs/detailed'
```

```python
#Define path where webdriver was downloaded
path = "/Users/ismaelcisse/Downloads/chromedriver-mac-arm64/chromedriver"
```

```python
#For the current version of selenium, you can't pass executable_path to Chrome() directly. 
#If you want to reference a specific downloaded version of ChromeDriver try this:
cService = webdriver.ChromeService(path = "/Users/ismaelcisse/Downloads/chromedriver-mac-arm64/chromedriver")

#Define driver variable
driver = webdriver.Chrome(service = cService) #dont close that page

#Chrome because it works well on it (can be Safari,Mozilla etc...)
```

```python
#Request to the website
driver.get(website)
```

```python
#Locate an element
all_matches_button = driver.find_element(By.XPATH,'//label[@analytics-event="All matches"]')
#Using click feature (want to display all matches instead Home/Away split by default)
all_matches_button.click()
```

```python
#Extract data from a table
# table element 
# tr tag = table row
# td tag = table data
  
matches = driver.find_elements(By.TAG_NAME,'tr')

matches

date = []
home_team = []
score = []
away_team = []
for match in matches:
    date.append(match.find_element(By.XPATH,'./td[1]').text)
    home_team.append(match.find_element(By.XPATH,'./td[2]').text)
    score.append(match.find_element(By.XPATH,'./td[3]').text)
    away_team.append(match.find_element(By.XPATH,'./td[4]').text) 
```

```python
away_team
```

```python
#Exporting data to CSV files with Pandas

df = pd.DataFrame({'Date':date, "Home Team":home_team, "Score":score, "Away Team":away_team})
df
```
